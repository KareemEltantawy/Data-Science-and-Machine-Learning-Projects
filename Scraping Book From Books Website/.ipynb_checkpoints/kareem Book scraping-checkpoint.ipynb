```python
import requests
from bs4 import BeautifulSoup
import sqlite3
```

```python
response = requests.get('https://books.toscrape.com/catalogue/category/books/travel_2/index.html')
soup = BeautifulSoup(response.text, 'html.parser') # 'html.parser' ...> always use it
```

```python
connection = sqlite3.connect('books.db')

sql = "INSERT INTO books (title,price,rate) VALUES (?,?,?)" #names of columns must be like it in db browser for sqlite 


RATING_MAP = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5} # we will use it in the variable rate to convert three to 3
for article in soup.find_all('article', attrs={'class': 'product_pod'}):
    title = article.find('h3').find('a').attrs['title']
    price = float(article.find('p', attrs={'class': 'price_color'}).get_text().split('£')[-1])
    rate = RATING_MAP[article.find('p').attrs['class'][-1]]
    s
    data = [title,price,rate]
    
    connection.execute(sql, data)
    connection.commit()
    
connection.close()   
```

```python
import requests
import json
import sqlite3
```

```python
response = requests.get('https://dashboard.nbshare.io/api/v1/apps/reddit')
res = response.text
```

```python
type(res)
```

```python
raddit_data = json.loads(res)
raddit_data
```

```python
connection = sqlite3.connect('raddit.db') #make the database with db browser for sqlite before this step

sql = "INSERT INTO posts (no_of_comments,sentiment,sentiment_score,ticker) VALUES (?,?,?,?)" #names of columns must be like it in db browser for sqlite 

for raddit in raddit_data:
    no_of_comments = raddit['no_of_comments']
    sentiment = raddit['sentiment']
    sentiment_score = raddit['sentiment_score']
    ticker = raddit['ticker']
    
    
    data = [no_of_comments,sentiment,sentiment_score,ticker]
    
    connection.execute(sql, data)
    connection.commit()

connection.close()
```
